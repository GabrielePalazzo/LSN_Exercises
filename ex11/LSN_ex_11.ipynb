In this exercise I tried to change a few parameters ($N_{epochs}$, $N_{train}$ and $\sigma$) in order to see the dependence of the linear regression NN.

```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

titleSize = 20
axesLabelSize = 15
```

```python
def linearRegression(nSamples, nTrain, noise):
    m = 2
    b = 1

    np.random.seed(0)
    x_train = np.random.uniform(-1, 1, nSamples)
    x_valid = np.random.uniform(-1, 1, 10)
    x_valid.sort()
    y_target = m * x_valid + b # ideal (target) linear function

    sigma = noise # noise standard deviation, for the moment it is absent
    y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
    y_valid = np.random.normal(m * x_valid + b, sigma)

    model = tf.keras.Sequential()
    model.add(Dense(1, input_shape=(1,)))

    # compile the model choosing optimizer, loss and metrics objects
    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
    #model.summary()

    history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=nTrain, verbose=0,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
    model.get_weights()
    # evaluate model
    score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=0)

    # print performance
    #print()
    #print('Test loss:', score[0])
    #print('Test accuracy:', score[1])
    # evaluate model with the exact curve
    score = model.evaluate(x_valid, y_target, batch_size=32, verbose=0)

    
    x_predicted = np.random.uniform(-1, 1, 100)
    y_predicted = model.predict(x_predicted)
    
    return score[0], model.get_weights()[0], model.get_weights()[1]
    
```

First of all I changed $N_{train}$ ($N_{epochs} = 30$, $\sigma = 0.1$):

```python
# (1)

a = 200
b = 2000
density = 50
score = []
M = []
B = []
for i in range(density):
    newScore, newM, newB = linearRegression(int(i*((b-a)/density) + a),30,0.1)
    score.append(newScore)
    M.append(newM[0])
    B.append(newB[0])
    #print(newM[0], newB)


xPlot = []
for i in range(density):
    xPlot.append(int(i*(b-a)/density + a))
    
yPlot = []
for i in range(density):
    yPlot.append(score[i])

plt.figure(figsize=(12, 12))
plt.title('Model loss', fontsize=titleSize)
plt.plot(xPlot, yPlot)
plt.ylabel('Loss', fontsize=axesLabelSize)
plt.xlabel('$N_{train}$', fontsize=axesLabelSize)
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 12))
plt.title('$m$ and $b$', fontsize=titleSize)
plt.plot(xPlot, M, 'r-', label='m')
plt.plot(xPlot, B, 'g-', label='b')
plt.plot(xPlot, 2 + np.zeros(len(xPlot)), 'r--', label='ideal m')
plt.plot(xPlot, 1 + np.zeros(len(xPlot)), 'g--', label='ideal b')
plt.legend(loc='best')
plt.ylabel('$m$, $b$', fontsize=axesLabelSize)
plt.xlabel('$N_{train}$', fontsize=axesLabelSize)
plt.grid(True)
plt.show()
```

The loss decays with $N_{train} > 1000$. Encreasing $N_{train}$ even further produces no significant result.

I need a more complex model (with more parameters).

The number of parameters is equal to:
$$
N_{param} = 2n_0 + \sum_{i=2}^{N_{layers}} n_i \cdot (n_{i-1} + 1)
$$
where $n_i$ is the number of neurons in the $i$-th layer.

```python
def cubicRegression(nSamples, nTrain, noise, layerNumber, neuronsPerLayer, Activation, Optimizer, lossFunction, Range, Label):
    # a x^3 + b x^2 + c x + d
    a =  3
    b = -2
    c = -3
    d =  4
    
    inputParameterNumber = 1

    np.random.seed(0)
    x_train = np.random.uniform(-1, 1, nSamples)
    x_valid = np.random.uniform(-1, 1, 100)
    x_valid.sort()
    
    y_target = a * x_valid**3 + b * x_valid**2 + c * x_valid + d # ideal (target) linear function
    
    sigma = noise # noise standard deviation, for the moment it is absent
    y_train = np.random.normal(a * x_train**3 + b * x_train**2 + c * x_train + d, sigma) # actual measures from which we want to guess regression parameters
    y_valid = np.random.normal(a * x_valid**3 + b * x_valid**2 + c * x_valid + d, sigma)

    model = tf.keras.Sequential()

    model.add(Dense(1, input_shape=(inputParameterNumber,)))
        #model.add(Dense(neuronsPerLayer, input_shape=(inputParameterNumber,)))
    for i in range(layerNumber - 2):
        model.add(Dense(neuronsPerLayer, activation=Activation))
    model.add(Dense(1, activation=Activation))

    # compile the model choosing optimizer, loss and metrics objects
    model.compile(optimizer=Optimizer, loss=lossFunction, metrics=['mse'])

    history = model.fit(x=x_train, y=y_train, batch_size=32, epochs=nTrain, verbose=0,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
    model.get_weights()
    # evaluate model
    score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=0)

    # print performance
    #print()
    #print('Test loss:', score[0])
    #print('Test accuracy:', score[1])
    # evaluate model with the exact curve
    score = model.evaluate(x_valid, y_target, batch_size=32, verbose=0)
    
    
    x_valid = np.random.uniform(Range[0], Range[1], 200)
    x_valid.sort()
    y_target = a * x_valid**3 + b * x_valid**2 + c * x_valid + d
    
    x1_predicted = np.random.uniform(-1, 1, 200)
    y1_predicted = model.predict(x1_predicted)
    x2_predicted = np.random.uniform(Range[0], Range[1], 200)
    y2_predicted = model.predict(x2_predicted)
    #plt.figure(figsize=(10, 10))
    #plt.title('$f(x)$', fontsize=titleSize)
    plt.subplot(121)
    plt.scatter(x1_predicted, y1_predicted, label=Label) #color='r')
    plt.subplot(122)
    plt.scatter(x2_predicted, y2_predicted, label=Label)
    #plt.plot(x_valid, y_target)
    #plt.grid(True)
    #plt.ylabel('$f(x)$', fontsize=axesLabelSize)
    #plt.xlabel('$x$', fontsize=axesLabelSize)
    #plt.show()
    return score[0]
```
